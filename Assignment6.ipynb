```python
df1 = df1.replace(',','', regex=True)
convert_dict = {'2018 Mean Annual Wage': int}
df1=df1.astype(convert_dict)
df1.dtypes
```

```python
df1=df1.rename(columns={"Sate ":"State","2018 Mean Annual Wage": "Mean_annual_wage"})
df1.head()
```
